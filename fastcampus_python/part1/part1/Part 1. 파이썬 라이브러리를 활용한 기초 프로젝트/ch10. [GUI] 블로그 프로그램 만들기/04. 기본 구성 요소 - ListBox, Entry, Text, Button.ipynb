#### 10-4-1. ë¦¬ìŠ¤íŠ¸ ë°•ìŠ¤ (ListBox)
- ë¦¬ìŠ¤íŠ¸ë°•ìŠ¤ëŠ” **ì •í•´ì§„ ìˆœì„œê°€ ìˆëŠ” ì—¬ëŸ¬ ê°œì˜ ë°ì´í„°ë¥¼ í‘œì‹œ**í•˜ëŠ” ì»´í¬ë„ŒíŠ¸

```python
from tkinter import *

root = Tk()                                # tkinter ê°ì²´(ì°½) ìƒì„±
root.geometry('200x200')
listbox = Listbox(root)                    # root ì°½ì— Listbox ì»´í¬ë„ŒíŠ¸(ìœ„ì ¯) ì¶”ê°€
listbox.pack()                             # Listbox ê°ì²´ë¥¼ ì°½ì— í‘œì‹œ

for i in ['ì²«ë²ˆì§¸ ìš”ì†Œ', 'ë‘ë²ˆì§¸ ìš”ì†Œ', 'ì„¸ë²ˆì§¸ ìš”ì†Œ', 'ë„¤ë²ˆì§¸ ìš”ì†Œ']:
    listbox.insert(END, i)                 # Listbox ë§ˆì§€ë§‰ ìœ„ì¹˜ì— ìƒˆë¡œìš´ ë°ì´í„° ì¶”ê°€

root.mainloop()                            # root ì°½ì„ ë©”ì¸ ë£¨í”„ì— íƒœìš°ê¸°
                                           # ì¢…ë£Œë˜ì§€ ì•Šê³  ì´ë²¤íŠ¸ ìˆ˜ì‹  ë“±ì˜ ì—­í• ì„ ìˆ˜í–‰
```

```python
from tkinter import *

def event_for_listbox(event):
    print("ì„ íƒë˜ì—ˆìŠµë‹ˆë‹¤ !")

root = Tk()                                # tkinter ê°ì²´(ì°½) ìƒì„±
root.geometry('200x200')
listbox = Listbox(root)                    # root ì°½ì— Listbox ì»´í¬ë„ŒíŠ¸(ìœ„ì ¯) ì¶”ê°€
listbox.pack()                             # Listbox ê°ì²´ë¥¼ ì°½ì— í‘œì‹œ

for i in ['ì²«ë²ˆì§¸ ìš”ì†Œ', 'ë‘ë²ˆì§¸ ìš”ì†Œ', 'ì„¸ë²ˆì§¸ ìš”ì†Œ', 'ë„¤ë²ˆì§¸ ìš”ì†Œ']:
    listbox.insert(END, i)                 # Listbox ë§ˆì§€ë§‰ ìœ„ì¹˜ì— ìƒˆë¡œìš´ ë°ì´í„° ì¶”ê°€

listbox.bind('<<ListboxSelect>>', event_for_listbox) # ë¦¬ìŠ¤íŠ¸ ë°•ìŠ¤ ì„ íƒ ì‹œ event_for_listbox í•¨ìˆ˜ í˜¸ì¶œ
    
    
root.mainloop()                            # root ì°½ì„ ë©”ì¸ ë£¨í”„ì— íƒœìš°ê¸°
                                           # ì¢…ë£Œë˜ì§€ ì•Šê³  ì´ë²¤íŠ¸ ìˆ˜ì‹  ë“±ì˜ ì—­í• ì„ ìˆ˜í–‰
```

#### 10-4-2. ì—”íŠ¸ë¦¬ (Entry)

- ì‚¬ìš©ì ì…ë ¥ì— ë”°ë¥¸ ì´ë²¤íŠ¸ ìƒì„± ë²„ì „

```python
from tkinter import *

def event_for_entry(event):
    print(f"ì‚¬ìš©ì ì…ë ¥ ê°’ì€ {entry.get()} ì…ë‹ˆë‹¤ !")  

root = Tk()
root.geometry('200x200')

entry = Entry(root)                          # root ì°½ì— Listbox ì»´í¬ë„ŒíŠ¸(ìœ„ì ¯) ì¶”ê°€
entry.insert(0, "ë°ì´í„°ë¥¼ ì…ë ¥í•´ ì£¼ì„¸ìš”")    # ì²«ë²ˆì§¸ ë°ì´í„°ì— ì—”íŠ¸ë¦¬ ì…ë ¥ê°’ ì €ì¥
entry.bind("<Return>",event_for_entry)       # ì—”íŠ¸ë¦¬ ê°’ Return ì‹œ event_for_entry í•¨ìˆ˜ í˜¸ì¶œ
entry.pack()                                 # ì—”íŠ¸ë¦¬ ê°ì²´ë¥¼ ì°½ì— í‘œì‹œ

root.mainloop()
```
