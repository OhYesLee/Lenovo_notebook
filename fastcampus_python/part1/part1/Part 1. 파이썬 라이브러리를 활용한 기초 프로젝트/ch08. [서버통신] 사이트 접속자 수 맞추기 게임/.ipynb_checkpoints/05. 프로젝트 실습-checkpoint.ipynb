#### 8-5-1. κ²μ„ μ„λ²„ λ§λ“¤κΈ°

```python
# multi_play_game_server.py

import socket
import select
import random

HOST = 'localhost'   # μ ‘μ† μ„λ²„ μ£Όμ†
PORT = 50007         # ν΄λΌμ΄μ–ΈνΈ μ ‘μ†μ„ λ€κΈ°ν•λ” ν¬νΈλ²νΈ

with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:    # μ†μΌ“ μƒμ„±
    s.bind((HOST, PORT))   # μ†μΌ“κ³Ό ν¬νΈ μ—°κ²°
    s.listen()             # μ†μΌ“ listening μ‹μ‘
    print('μ„λ²„κ°€ μ‹μ‘λμ—μµλ‹λ‹¤.')

    readsocks = [s]    # select ν•¨μμ—μ„ κ΄€μ°°λ  μ†μΌ“ λ¦¬μ¤νΈ μƒμ„±
    answers = {}       # μ •λ‹µ λ”•μ…”λ„λ¦¬ μƒμ„± (ν”λ μ΄μ–΄ λ³„ μ •λ‹µ μ €μ¥)
    num_clients = 0    # ν„μ¬ μ ‘μ†μ μ

    while True:
        # select ν•¨μλ΅ μ†μΌ“λ“¤μ μƒνƒλ¥Ό ν™•μΈ
        readables, writeables, exceptions = select.select(readsocks, [], [])
        
        for sock in readables:
            if sock == s:  # μ‹ κ· ν΄λΌμ΄μ–ΈνΈ μ ‘μ†
                newsock, addr = s.accept()
                num_clients += 1  # μ ‘μ†μ μ μ¦κ°€
                print(f'ν΄λΌμ΄μ–ΈνΈκ°€ μ ‘μ†ν–μµλ‹λ‹¤:{addr}, μ •λ‹µμ€ {num_clients} μ…λ‹λ‹¤.')
                readsocks.append(newsock)
                answers[newsock] = num_clients  # μ‹ κ· ν΄λΌμ΄μ–ΈνΈ μ •λ‹µ μƒμ„±
                for key in answers.keys():      # κΈ°μ΅΄ ν΄λΌμ΄μ–ΈνΈ μ •λ‹µ μ—…λ°μ΄νΈ
                    answers[key] = num_clients
                    
            else:  # μ΄λ―Έ μ ‘μ†ν• ν΄λΌμ΄μ–ΈνΈμ μ”μ²­ (κ²μ„μ§„ν–‰μ„ μ„ν• μ”μ²­)
                conn = sock
                data = conn.recv(1024).decode('utf-8')
                print(f'λ°μ΄ν„°:{data}')

                try:  # μ«μλ¥Ό μ…λ ¥ν•μ§€ μ•μ„ κ²½μ° μ‘λ‹µ λ©”μ„Έμ§€
                    n = int(data)
                except ValueError:
                    conn.sendall(f'μ…λ ¥κ°’μ΄ μ¬λ°”λ¥΄μ§€ μ•μµλ‹λ‹¤:{data}'.encode('utf-8'))
                    continue
                
                # μ •λ‹µ κ°€μ Έμ¤κΈ°
                answer = answers.get(conn)
                
                if n == 0:  # ν΄λΌμ΄μ–ΈνΈκ°€ μΆ…λ£ 0 λ¥Ό μ…λ ¥ν•  κ²½μ°
                    conn.sendall(f"μΆ…λ£".encode('utf-8'))
                    conn.close()
                    num_clients -= 1  # μ ‘μ†μ μ κ°μ†
                    print(f'ν΄λΌμ΄μ–ΈνΈκ°€ κ²μ„μ„ μΆ…λ£ν–μµλ‹λ‹¤:{addr}, μ •λ‹µμ€ {num_clients} μ…λ‹λ‹¤.')
                    
                    for key in answers.keys():  # κΈ°μ΅΄ ν΄λΌμ΄μ–ΈνΈ μ •λ‹µ μ—…λ°μ΄νΈ
                        answers[key] = num_clients
                    readsocks.remove(conn)  # ν΄λΌμ΄μ–ΈνΈ μ ‘μ† ν•΄μ μ‹ readsocksμ—μ„ μ κ±°
                
                # ν΄λΌμ΄μ–ΈνΈμ μ…λ ¥κ°’ μ±„μ 
                elif n > answer:
                    conn.sendall("λ„λ¬΄ λ†’μ•„μ”".encode('utf-8'))
                elif n < answer:
                    conn.sendall("λ„λ¬΄ λ‚®μ•„μ”".encode('utf-8'))
                else:
                    conn.sendall("μ •λ‹µμ…λ‹λ‹¤ !".encode('utf-8'))
```

#### 8-5-2. ν΄λΌμ΄μ–ΈνΈ λ§λ“¤κΈ°

```python
# multi_play_game_client.py

import socket

HOST = 'localhost'   # μ ‘μ† μ„λ²„ μ£Όμ†
PORT = 50007         # ν΄λΌμ΄μ–ΈνΈ μ ‘μ†μ„ λ€κΈ°ν•λ” ν¬νΈλ²νΈ

with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:    # μ†μΌ“ μƒμ„±
    s.connect((HOST, PORT))  # μ„λ²„ μ—°κ²°

    while True:
        # μ •λ‹µ μ…λ ¥
        n = input("μμƒ ν”λ μ΄μ–΄ μ«μλ¥Ό μ…λ ¥ν•μ„Έμ”(0μ€ κ²μ„ν¬κΈ°):")
        
        # μ…λ ¥κ°’μ΄ μ—†μ„ κ²½μ°
        if not n.strip():
            print("μ…λ ¥κ°’μ΄ μλ»λμ—μµλ‹λ‹¤.")
            continue
        
        # μ„λ²„ μ†΅μ‹  (μΈμ½”λ”©)
        s.sendall(n.encode('utf-8'))
        
        # μ„λ²„ μ‘λ‹µ μμ‹  (λ””μ½”λ”©)
        data = s.recv(1024).decode('utf-8')
        
        # μ„λ²„ μ‘λ‹µμ΄ 'μ •λ‹µ'μ΄κ±°λ‚ 'μΆ…λ£' μ΄λ©΄ κ²μ„ μΆ…λ£
        print(f'μ„λ²„μ‘λ‹µ:{data}')
        if data == "μ •λ‹µ" or data == "μΆ…λ£":
            break
```
