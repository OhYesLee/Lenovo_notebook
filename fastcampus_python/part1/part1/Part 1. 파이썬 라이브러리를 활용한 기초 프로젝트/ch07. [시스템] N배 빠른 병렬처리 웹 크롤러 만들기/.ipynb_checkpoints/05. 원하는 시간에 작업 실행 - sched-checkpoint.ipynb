#### 7.5.1. sched ì˜ˆì œ
â‘  í”„ë¡œê·¸ë¨ ì‹¤í–‰ í›„ 5ì´ˆ í›„ì— print_a() í˜¸ì¶œ <br>
â‘¡ í”„ë¡œê·¸ë¨ ì‹¤í–‰ í›„ 3ì´ˆ í›„ì— print_b() í˜¸ì¶œ <br>
â‘¢ í”„ë¡œê·¸ë¨ ì‹¤í–‰ í›„ 7ì´ˆ í›„ì— print_c() í˜¸ì¶œ

```python
import sched
import time

start = time.time()

def print_a(a):
    print(round(time.time() - start, 2), ' ì´ˆ ê²½ê³¼')
    print(a)


def print_b(b):
    print(round(time.time() - start, 2), ' ì´ˆ ê²½ê³¼')
    print(b)


def print_c(c):
    print(round(time.time() - start, 2), ' ì´ˆ ê²½ê³¼')
    print(c)


s = sched.scheduler()  # ìŠ¤ì¼€ì¤„ëŸ¬ ê°ì²´ ìƒì„±
s.enter(5, 1, print_a, ('print_a í•¨ìˆ˜ ì‹¤í–‰ë¨',))  # 5ì´ˆ í›„ì— ì‹¤í–‰
s.enter(3, 1, print_b, ('print_b í•¨ìˆ˜ ì‹¤í–‰ë¨',))  # 3ì´ˆ í›„ì— ì‹¤í–‰
s.enter(7, 1, print_c, ('print_c í•¨ìˆ˜ ì‹¤í–‰ë¨',))  # 7ì´ˆ í›„ì— ì‹¤í–‰
s.run()
```

#### 7.5.6. ìŠ¤ì¼€ì¤„ëŸ¬ ì¤‘ì§€

```python
# ìŠ¤ì¼€ì¤„ëŸ¬ ì´ˆê¸°í™”
schedule.clear()
```

```python
# job ì„¤ì •
job = schedule.every(1).seconds.do(message, '1ì´ˆ')  # ì´ë²¤íŠ¸ ë“±ë¡
```

```python
# ìŠ¤ì¼€ì¤„ëŸ¬ ì‹¤í–‰
count = 0

while True:
    schedule.run_pending()
    time.sleep(1)
    
    count = count + 1
    
    if count > 5:  # 5íšŒ ì‹¤í–‰ í›„ ìŠ¤ì¼€ì¤„ëŸ¬ ì¤‘ì§€
        schedule.cancel_job(job)
        print('ìŠ¤ì¼€ì¤„ëŸ¬ê°€ ì¢…ë£Œë˜ì—ˆìŠµë‹ˆë‹¤ !')
        break
```
